```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("google/pegasus-cnn_dailymail")
```

```python
def convert_examples_to_features(example_batch):
    input_encodings = tokenizer(
        [str(d) for d in example_batch['dialogue']],
        max_length=1024,
        truncation=True,
        padding=True
    )

    target_encodings = tokenizer(
        [str(s) for s in example_batch['summary']],
        max_length=128,
        truncation=True,
        padding=True
    )

    return {
        'input_ids': input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': target_encodings['input_ids']
    }
```

```python
##model evalutaion
def generate_batch_sized_chunks(list_of_elem,batch_size):
  for i in range(0, len(list_of_elem), batch_size):
    yield list_of_elem[i:i+ batch_size]
```

```python
def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                                 batch_size=16, device='cuda',
                                 column_text='article',
                                 column_summary='highlights'):
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for article_batch, target_batch in zip(article_batches, target_batches):
        inputs = tokenizer(article_batch, truncation=True,
                          padding='max_length', return_tensors='pt').to(device)
        summaries = model.generate(inputs['input_ids'])
        decoded = tokenizer.batch_decode(summaries, skip_special_tokens=True)
        metric.add_batch(predictions=decoded, references=target_batch)

    return metric.compute()
```
